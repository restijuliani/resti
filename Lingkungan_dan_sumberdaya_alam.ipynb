```python
import pandas as pd

path='/content/drive/MyDrive/CSV/indeks-standar-pencemar-udara-di-provinsi-dki-jakarta-bulan-april-tahun-2020.csv'
df = pd.read_csv(path)
df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt # 

df.plot(kind='scatter', x='o3', y='max') 

plt.title('indeks data penyebaran udara bulan april 2020')
plt.xlabel('o3')
plt.ylabel('max')
plt.xlim(0,150)
plt.ylim(0,150)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array (df['o3'])
y = np.array (df['max'])
print(f'x : {x}')
print(f'y : {y}')
```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()
model.fit(x, y)
```

```python
x_vis =np.array([0,10000]).reshape(-1, 1)
y_vis =model.predict(x_vis)
```

```python
plt.scatter(x, y)
plt.plot(x_vis, y_vis, '-r' )
plt.title('indeks data penyebaran udara bulan april 2020')
plt.xlabel('o3')
plt.ylabel('max')
plt.xlim(0,130)
plt.ylim(0,150)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
balita = df['o3']
daftar = df['max']
c= ['red', 'green', 'blue']
plt.title(" bar")
plt.bar(balita, daftar, color= c)
plt.show()
```

```python
import matplotlib.pyplot as plt
# pe chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
nama= df['o3']
hasil = df['max']
c= ['red', 'green', 'blue']
ax.pie(nama,labels=hasil,autopct='%1.2ff%%')
plt.title(" pie chat")
plt.show()
```
