```python
import pandas as pd

path='/content/drive/MyDrive/CSV/tahun-2018-data-posyandu.csv'
df = pd.read_csv(path)
df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt # 

df.plot(kind='scatter', x='Jumlah Balita Terdaftar', y='Jumlah Balita Aktif') 

plt.title('indeks data jumlah Registrasi penduduk')
plt.xlabel('Jumlah Balita Terdaftar')
plt.ylabel('Jumlah Balita Aktif')
plt.xlim(0,70)
plt.ylim(0,70)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array (df['Jumlah Balita Terdaftar'])
y = np.array (df['Jumlah Balita Aktif'])
print(f'x : {x}')
print(f'y : {y}')
```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()
model.fit(x, y)
```

```python
x_vis =np.array([0,10000]).reshape(-1, 1)
y_vis =model.predict(x_vis)
```

```python
plt.scatter(x, y)
plt.plot(x_vis, y_vis, '-r' )
plt.title('indeks data jumlah posyandu 2018')
plt.xlabel('Jumlah Balita Terdaftar')
plt.ylabel('Jumlah Balita Aktif')
plt.xlim(0,70)
plt.ylim(0,70)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
balita = df['Jumlah Balita Terdaftar']
daftar = df['Jumlah Balita Aktif']
c= ['red', 'green', 'blue']
plt.title(" bar")
plt.bar(balita, daftar, color= c)
plt.show()
```

```python
import matplotlib.pyplot as plt
# pe chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
balita = df['Jumlah Balita Terdaftar']
daftar = df['Jumlah Balita Aktif']
c= ['red', 'green', 'blue']
ax.pie(daftar,labels=balita,autopct='%1.2ff%%')
plt.title(" pie chat")
plt.show()
```
